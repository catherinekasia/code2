# Density based clustering

* Step 1: create a box graph. This part is already done for you. To see what exactly happens, take a look at the ```boxgraph.py``` file.
* Step 2: find the core points. You are expected to implement ```find_core_points()``` as found below and use it to assign all cluster points that are core points the proper function.
* Step 3: compute the cluster cores. You are expected to implement ```compute_cluster_cores()``` as found below. Here, you are expected to give all core points that belong to the same cluster the same label, using a unique label for each core.
* Step 4: assign border points to the correct cluster. You are expected to implement ```assign_border_points()```: assign all cluster points that are border points the label of the closest core point.

* Step 2: assign all cluster points that are core points the CORE functionality. For more on this, see the ```ClusterPoint``` class in ```point.py```. 

````python
def find_core_points(bg):
    """
    Step 2: find the core points
    """

#// BEGIN_TODO [FIND-CORE-POINTS]

    for box in bg.boxes:
        for p in box.points:
            count = 0
            for nb in [box] + box.neighbours:
                for q in nb.points:
                    if p.sq_distance_to(q) <= bg.eps_sqr:
                        count += 1
            if count >= bg.min_pts:
                p.func = CP.Func.CORE

#// END_TODO [FIND-CORE-POINTS]                    
````

* Step 3: compute the cluster cores and label the points accordingly.

````python
def compute_cluster_cores(bg):
    """
    Step 3: compute the cluster cores    
    """
    nr_clusters = -1

#// BEGIN_TODO [COMPUTE-CLUSTER-CORES]

    for box in bg.boxes:
        has_core = any(p.func == CP.Func.CORE for p in box.points)
        if has_core and not box.is_labeled():
            nr_clusters += 1
            queue = [box]
            box.label = nr_clusters
            while queue:
                current = queue.pop(0)
                for neighbour in current.neighbours:
                    if not neighbour.is_labeled() and current.is_core_neighbour(neighbour, bg.eps):
                        neighbour.label = nr_clusters
                        queue.append(neighbour)

    for box in bg.boxes:
        if box.is_labeled():
            for p in box.points:
                if p.func == CP.Func.CORE:
                    p.cluster_label = box.label

#// END_TODO [COMPUTE-CLUSTER-CORES]
    return nr_clusters
````

* Step 4: assign border points to the closest cluster by giving them the same label as the other points in that cluster.

````python
def assign_border_points(bg):
    """
    Step 4: assign border points to clusters
    """
#// BEGIN_TODO [ASSIGNING-BORDER-POINTS]

    for box in bg.boxes:
        for p in box.points:
            if p.func != CP.Func.CORE:
                min_sq_dist = float('inf')
                closest_label = -1
                for nb in [box] + box.neighbours:
                    for q in nb.points:
                        if q.func == CP.Func.CORE:
                            sq_d = p.sq_distance_to(q)
                            if sq_d <= bg.eps_sqr and sq_d < min_sq_dist:
                                min_sq_dist = sq_d
                                closest_label = q.cluster_label
                if closest_label != -1:
                    p.func = CP.Func.BORDER
                    p.cluster_label = closest_label

#// END_TODO [ASSIGNING-BORDER-POINTS]
````

````python
def run(path_in, path_out):
    """
    Reads the input set, clusters the points and writes to output

    :param path_in:     location of the input set
    :param path_out:    location to print the output
    """
    # Step 1: read file and create box graph
    try:
        with open(path_in, "r") as f:
            bg = BoxGraph.read_input(f)    
    except IOError:
        print("could not read input file:", path_in, file=sys.stderr)
        return
    
    # Step 2: Compute the core points
    find_core_points(bg)

    # Step 3: cluster the core points
    nr_clusters = compute_cluster_cores(bg)

    # Step 4: assign border points to clusters
    assign_border_points(bg)
    
    try:
        bg.write_output(nr_clusters, path_out, assignment_nr)
    except IOError:
        print("could not write output to file:", path_out, file=sys.stderr)
    print("Cluster counter:", nr_clusters, file=sys.stderr)
````
